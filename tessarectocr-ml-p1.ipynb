```python
import pytesseract
from PIL import Image
from pdf2image import convert_from_path

images = convert_from_path('/kaggle/input/ml-practicals/sample.pdf')
for i in range(len(images)):
  images[i].save('page' + str(i) + '.jpg', 'JPEG')
  print("Image: "+ str(i))
  print(pytesseract.image_to_string(Image.open('page' + str(i) + '.jpg')))
```
